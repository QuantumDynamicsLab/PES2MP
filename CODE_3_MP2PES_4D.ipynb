```python
import numpy as np
import pandas as pd
```

#### Import V_lambdas

```python
df_Vn = pd.read_csv('4D_Vlam_6_4.dat',sep=',')  
```

```python
df_Vn
```

#### Declare important variables and import Legendre Coeff numpy file

```python
lm=28                        # Radial terms
nc=77                      # number of Radial coordinates (Must be same for all angles)
ngm = 112                    # number of angular coordinates
E_inf = -78.70465827  
px = np.zeros((ngm,lm))     # Matrix to stores legendre coeffinients

V_nf= np.zeros((nc,lm))     # Numpy 2D array to store generated potentials
V_nf2= np.zeros((ngm,nc))   # Numpy 2D array to store ab_initio potentials
```

```python
px = np.load("4D_BiSp_coeff.npy")
V_n1 = px.dot(df_Vn.T)                 # save regenerated potentials
```

#### convert regenerated potentials into dataframe

```python
a12 = np.arange(ngm)            # creates header for angles
df_Vnf = pd.DataFrame(V_n1.T, columns = a12) # saves final matrix into dataframe with appropriate header
df_Vnf                     # prints regenerated potentials
```

#### import ab initio potentials --> convert into dataframe --> find error

```python
df_inp2 = pd.read_csv('4D_PES.dat',header=None,sep='\s+')       # importing ab initio data
df_inp2.sort_values(by = [0,1,2,3], inplace=True, ascending = True)
df_inp2.reset_index(inplace=True, drop = True)  # sorting by R, phi, th2 and th1 and reindexing
df_inp2
```

```python
np.shape(V_nf2)
```

```python
for i in range (nc):            # loop over all R   
    ct = i*ngm                  # extract start point. Since input dataframe is sorted by R and theta, 
    f = df_inp2[4][ct:ct+ngm]   # potentials (V) are extracting for each R value at a time 
    V_nf2[:,i] = f           # ab initio data stored in 2D matrix one by one
V_nf2                        # visualizing numoy file
```

```python
df_Vnf2 = pd.DataFrame(V_nf2.T, columns = a12) # saves final matrix into dataframe with angular header
df_Vnf2
```

#### Warning! : Recreating 91 ab-initio angles from 4 radial coefficients. Ideally,
    {no of angular terms = no of V lambdas}

```python
# Error at minima (approx location)
st, end = 25, 35
rmse_min = np.sqrt(np.mean(np.square(df_Vnf[st:end]-df_Vnf2[st:end])))
rmse_min.mean()
```

```python
# Error at high energy (approx location)
end = 25
rmse_min = np.sqrt(np.mean(np.square(df_Vnf[:end]-df_Vnf2[:end])))
rmse_min.mean()
```

```python
# Error at asymptotic (approx location)
st = 35
rmse_min = np.sqrt(np.mean(np.square(df_Vnf[st:]-df_Vnf2[st:])))
rmse_min.mean()
```
